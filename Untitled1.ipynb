```python
#import facebooks prophet plotting library
from fbprophet.plot import plot_plotly
from fbprophet import Prophet
#Create the Prophet Object (model)
fbp = Prophet(daily_seasonality = True)
#fit/Train the model
fbp.fit(df)
future = fbp.make_future_dataframe(periods = 365)
forecast = fbp.predict(future)
#plot the data
plot_plotly(fbp, forecast)
```

```python
#Show the models prediction for 2010-03-29
forecast[forecast.ds=='2010-03-29']['yhat']
```
